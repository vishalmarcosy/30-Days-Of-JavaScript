```python
#aks the user to guess a number
#input by user has to be matched by number generated by computer
#if matched then correct 
#if not then ask again(loops-3 times)
#at the end will display the correct number
```

```python
import random
number =random.randint(1,10)
for i in range(0,3):
 user=int(input("Guess the lucky number"))
if user ==number:
  print("Hurray")
  print(f"You guessed the number right. The answer is {number}")
  break

if user!=number:
  print(f"Your guess is incorrect. The number is {number}")
```
